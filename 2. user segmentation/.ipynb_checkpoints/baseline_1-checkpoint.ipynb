```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate, train_test_split, StratifiedShuffleSplit
from sklearn import linear_model, metrics
from sklearn.linear_model import SGDClassifier as SGD
```

```python
train_directory = "data/train.csv"

df = pd.read_csv(train_directory, header=0 )

df['y'] = [1 if y == 'yes' else 0 for y in df['y']]
```

```python
pdays_mean_lt999 = df['y'][df['pdays'] != 999].mean()
pdays_mean_eq999 = df['y'][df['pdays'] == 999].mean()

print(f"Mean y [pdays != 999] = {pdays_mean_lt999}")
print(f"Mean y [pdays == 999] = {pdays_mean_eq999}")

df['pdays_eq999'] = df['pdays'].apply(lambda x: 0 if x == 999 else 1)
# pdays != 999 are more often overdue payment
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

from sklearn.linear_model import LogisticRegression

categorical_features = ['job', 'default', 'education', 'month',  'poutcome']
numerical_features = ['age', 'previous', 'duration', 'campaign', 'pdays_eq999', 'is_single', 
                      'is_cellular', 'emp.var.rate', 'euribor3m', 'nr.employed']

```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

X = df[categorical_features + numerical_features]
y = df['y']

scores = []
spliter = StratifiedShuffleSplit(n_splits=5, test_size=0.3)

for train_index, test_index in spliter.split(X, y):
    clf = RandomForestClassifier(n_estimators=20)

    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    preprocessor = ColumnTransformer(transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

    X_preproc = preprocessor.fit_transform(X_train)

    clf.fit(X_preproc, y_train)
    y_predicted = clf.predict(preprocessor.transform(X_test))

    score = roc_auc_score(y_test, y_predicted)
    scores.append(score)
roc_auc_mean = np.mean(scores)

# optimal depth 10

print(f"ROC AUC mean: {roc_auc_mean}")
```
